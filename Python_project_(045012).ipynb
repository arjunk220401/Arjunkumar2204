```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt

# Define the URL and the table class
wikiurl = "https://en.m.wikipedia.org/wiki/FIFA_World_Cup"
table_class = "wikitable sortable jquery-tablesorter"

# Get the response in the form of HTML
response = requests.get(wikiurl)

# Parse data from the HTML into a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')

# Find all tables on the page
tables = soup.find_all('table', {'class': 'wikitable'})

# Select the second table (index 1) from the list of tables
second_table = tables[1]

# Read the table data into a DataFrame and convert 'Year' to integers, skipping the header row
df = pd.read_html(str(second_table), header=0, converters={'Year': lambda x: int(x) if x.isnumeric() else 0})[0]

# Filter the DataFrame to include only rows where 'Year' is between 1930 and 2022
df = df[(df['Year'] >= 1930) & (df['Year'] <= 2022)]

# Create a bar graph
plt.figure(figsize=(12, 6))
plt.bar(df['Year'], df['Matches'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Matches Played')
plt.title('FIFA World Cup Matches by Year (1930-2022)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# To find the maximum times a country has hosted the FIFA World Cup

import pandas as pd
import requests
from bs4 import BeautifulSoup

# Define the URL and the table class
wikiurl = "https://en.m.wikipedia.org/wiki/FIFA_World_Cup"
table_class = "wikitable sortable jquery-tablesorter"

# Get the response in the form of HTML
response = requests.get(wikiurl)

# Parse data from the HTML into a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')

# Find all tables on the page
tables = soup.find_all('table', {'class': 'wikitable'})

# Select the second table (index 1) from the list of tables
second_table = tables[1]

# Read the table data into a DataFrame
df = pd.read_html(str(second_table))[0]  # Convert to DataFrame

# Specify the MultiIndex column name for 'Hosts'
multiindex_column = ('Hosts', 'Hosts')

# Group the data by the MultiIndex column and count occurrences
host_counts = df.groupby(multiindex_column).size()

# Find the country(s) with the maximum count
max_host_countries = host_counts[host_counts == host_counts.max()]

# Print the country(s) that hosted the maximum number of times
print("Country(s) that hosted the maximum number of times:")
print(max_host_countries)

#This code will output the country(s) that have hosted the FIFA World Cup the maximum number of times based on the data in the MultiIndex DataFrame.












```

```python
# To find the continents of all countries that have hosted the FIFA World Cup from 1930 to 2022, given a MultiIndex DataFrame,
#you can follow these steps:

#Create a dictionary that maps countries to continents.
#Filter the original DataFrame to get the years and hosting countries within the specified range (1930 to 2022).
#Use the 'Hosts' column from the MultiIndex DataFrame to identify the hosting countries.
#Map the hosting countries to continents using the dictionary.

import pandas as pd
import requests
from bs4 import BeautifulSoup
import re

# Define the URL and the table class
wikiurl = "https://en.m.wikipedia.org/wiki/FIFA_World_Cup"
table_class = "wikitable sortable jquery-tablesorter"

# Get the response in the form of HTML
response = requests.get(wikiurl)

# Parse data from the HTML into a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')

# Find all tables on the page
tables = soup.find_all('table', {'class': 'wikitable'})

# Select the second table (index 1) from the list of tables
second_table = tables[1]

# Read the table data into a DataFrame
df = pd.read_html(str(second_table), header=[0, 1])[0]  # Convert to MultiIndex DataFrame

# Define a function to extract the year from the 'Year' column
def extract_year(value):
    match = re.search(r'\d{4}', str(value))
    if match:
        return int(match.group())
    else:
        return None

# Apply the extract_year function to the 'Year' column
df[('Year', 'Year')] = df[('Year', 'Year')].apply(extract_year)

# Filter the DataFrame to get the years and hosting countries from 1930 to 2022
filtered_df = df[(df[('Year', 'Year')] >= 1930) & (df[('Year', 'Year')] <= 2022)]

# Extract the hosting countries and their respective years
hosting_info = filtered_df[[
    ('Hosts', 'Hosts'),
    ('Year', 'Year')
]].dropna()

# Create a dictionary that maps countries to continents (example mapping)
country_to_continent = {
    'Uruguay': 'South America',
    'Italy': 'Europe',
    'France': 'Europe',
    'Brazil': 'South America',
    'Switzerland': 'Europe',
    'Sweden': 'Europe',
    'Chile': 'South America',
    'England': 'Europe',
    'Mexico': 'North America',
    'West Germany': 'Europe',
    'Argentina': 'South America',
    'Italy[b]': 'Europe',  # Special handling for Italy with '[b]'
    'Spain': 'Europe',
    'United States': 'North America',
    'South Korea': 'Asia',
    'Japan':'Asia',
    'Germany': 'Europe',
    'South Africa': 'Africa',
    'Brazil': 'South America',
    'Russia': 'Europe',
    'Qatar': 'Asia',
}

# Map hosting countries to continents using the dictionary
hosting_info['Continent'] = hosting_info[('Hosts', 'Hosts')].map(country_to_continent)

# Print the DataFrame with countries, years, and continents
print(hosting_info)
```

```python
# Now we do the Regression Analysis - To show the relationship between 'matches' and 'year' in our data.

import pandas as pd
import requests
from bs4 import BeautifulSoup
import statsmodels.api as sm

# Define the URL and the table class
wikiurl = "https://en.m.wikipedia.org/wiki/FIFA_World_Cup"
table_class = "wikitable sortable jquery-tablesorter"

# Get the response in the form of HTML
response = requests.get(wikiurl)

# Parse data from the HTML into a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')

# Find all tables on the page
tables = soup.find_all('table', {'class': 'wikitable'})

# Select the second table (index 1) from the list of tables
second_table = tables[1]

# Read the table data into a DataFrame
df = pd.read_html(str(second_table), header=0)[0]  # Convert to DataFrame with header row

# Convert the 'Year' and 'Matches' columns to numeric types
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df['Matches'] = pd.to_numeric(df['Matches'], errors='coerce')

# Drop rows with missing data
df.dropna(subset=['Year', 'Matches'], inplace=True)

# Perform linear regression
X = sm.add_constant(df['Year'])  # Add a constant term to the model
y = df['Matches']

model = sm.OLS(y, X).fit()  # Fit the linear regression model

# Print the regression summary
print(model.summary())
```

```python
# Regression Analysis - show relationship between Total Attendance and Year.

import pandas as pd
import requests
from bs4 import BeautifulSoup
import statsmodels.api as sm

# Define the URL and the table class
wikiurl = "https://en.m.wikipedia.org/wiki/FIFA_World_Cup"
table_class = "wikitable sortable jquery-tablesorter"

# Get the response in the form of HTML
response = requests.get(wikiurl)

# Parse data from the HTML into a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')

# Find all tables on the page
tables = soup.find_all('table', {'class': 'wikitable'})

# Select the second table (index 1) from the list of tables
second_table = tables[1]

# Read the table data into a DataFrame
df = pd.read_html(str(second_table), header=0)[0]  # Convert to DataFrame with header row

# Convert the 'Year' and 'Total attendance †' columns to numeric types
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df['Total attendance †'] = pd.to_numeric(df['Total attendance †'], errors='coerce')

# Drop rows with missing data
df.dropna(subset=['Year', 'Total attendance †'], inplace=True)

# Perform linear regression
X = sm.add_constant(df['Year'])  # Add a constant term to the model
y = df['Total attendance †']

model = sm.OLS(y, X).fit()  # Fit the linear regression model

# Print the regression summary
print(model.summary())
```
